```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("../data/processed/ethiopia_fi_enriched.csv")


impact_data = {
    'Indicator': ['ACC_OWNERSHIP', 'ACC_MM_ACCOUNT', 'USG_DIGITAL_PAYMENT'],
    'Policy (NFIS-II)': [3.0, 1.5, 2.0],
    'Product (Telebirr)': [4.5, 12.5, 8.0],
    'Market Entry (M-Pesa)': [2.0, 10.0, 5.5],
    'Infrastructure (4G)': [5.0, 3.0, 4.5]
}

impact_matrix = pd.DataFrame(impact_data).set_index('Indicator')


plt.figure(figsize=(10, 6))
sns.heatmap(impact_matrix, annot=True, cmap="YlGnBu", fmt=".1f", cbar_kws={'label': 'Magnitude (pp boost)'})
plt.title("Task 3: Event-Indicator Association Matrix")
plt.xlabel("Event Category")
plt.ylabel("Financial Inclusion Indicator")
plt.savefig("../reports/figures/impact_association_matrix.png")
plt.show()
```
