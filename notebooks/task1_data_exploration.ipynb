```python
import pandas as pd
from datetime import datetime

def add_record(df, record, log_file="../data_enrichment_log.md"):
    """
    Safely add a new record to the DataFrame and log it.

    Parameters:
        df (pd.DataFrame): existing DataFrame
        record (dict): new observation/event/impact_link
        log_file (str): path to the enrichment log
    Returns:
        pd.DataFrame: updated DataFrame
    """

    record_id = record.get("record_id")
    if record_id in df['record_id'].values:
        print(f"Record {record_id} already exists. Skipping addition.")
        return df

    new_df = pd.DataFrame([record])
    df = pd.concat([df, new_df], ignore_index=True)
    print(f"Record {record_id} added.")

    today = datetime.today().strftime("%Y-%m-%d")
    log_entry = f"\n- **{today}** Added new record:\n"
    for k, v in record.items():
        log_entry += f"  - {k}: {v}\n"

    with open(log_file, "a") as f:
        f.write(log_entry)
    print(f"üìù Logged {record_id} in {log_file}.")

    return df
```
